### import libraries as basics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel('Data_Train.xlsx')
train_df.head()
```

```python
test_df = pd.read_excel('Test_set (1).xlsx')
test_df.head()
```

### for adding to datas use concat method

```python
final_df = pd.concat([train_df, test_df])
final_df.info()
```

```python
final_df.columns
```

### convert string date_of_journey to integer

```python
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df.drop('Date_of_Journey', axis = 1, inplace = True)
```

```python
final_df.head()
```

```python
final_df = final_df.reset_index()
```

```python
final_df.head()
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.head()
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
final_df.head()
```

```python
final_df['Arrival_Hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_Minutes'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head()
final_df['Arrival_Hour'] = final_df['Arrival_Hour'].astype(int)
final_df['Arrival_Minutes'] = final_df['Arrival_Minutes'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Arrival_Time',inplace = True , axis = 1)
```

```python
final_df['Dep_Hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_Minutes'] = final_df['Dep_Time'].str.split(':').str[1]
```

```python
final_df['Dep_Hour'] = final_df['Dep_Hour'].astype(int)
final_df['Dep_Minutes'] = final_df['Dep_Minutes'].astype(int)
```

```python
final_df.drop('Dep_Time',axis = 1, inplace = True)
```

```python
final_df.info()
```

#### Total Stops 


```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop' : 0, '1 stop' : 1, '2 stops' : 2, '3 stops': 3, '4 stops' : 4 ,'nan':1})
```

```python
final_df.head()
```

```python
final_df.drop('Route',axis = 1,inplace = True)
```

### duration

```python
final_df.head()
```

```python
final_df['Minutes'] = final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
final_df['Hours'] = final_df['Duration'].str.split(' ').str[0].str[0]
```

```python
final_df['Hours'] = final_df['Hours'].astype(int)
```

```python
final_df[final_df['Minutes'].isnull()] = 0
```

```python
final_df['Minutes'] = final_df['Minutes'].astype(int)
```

```python
final_df['Hours'] = final_df['Hours']*60
```

```python
final_df['Hours'].head()
```

```python
final_df['Duration'] = final_df['Hours'] + final_df['Minutes']
```

```python
final_df.drop('Hours',axis = 1,inplace = True)
final_df.drop('Minutes',axis = 1,inplace = True)
```

```python
final_df.head()
```
